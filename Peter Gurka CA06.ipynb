# Building Recommender System

```python
# make input variables for fitting
x = data[['IMDB Rating', 'Biography', 'Drama', 'Thriller', 'Comedy', 'Crime', 'Mystery', 'History']]
y = data['Movie Name']
```

```python
# Use Nearestneighbors and fit
neigh = NearestNeighbors(n_neighbors=5)
neigh.fit(x,y)
```

```python
# implement recommendation based on the movie 'post'
# IMDB Rating = 7.2, Biography = Yes, Drama = Yes, Thriller = No, Comedy = No, Crime = No, Mystery = No, History = Yes

rec_post = [7.2,1,1,0,0,0,0,1]
return_distance, z = neigh.kneighbors([rec_post])
print(z)
```

```python
# for loop to find 5 movies (aka output)

for five_recs in z:
    print(data.iloc[five_recs]['Movie Name'])
```
