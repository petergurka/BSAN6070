### Creating a year column
To view the trend over a period of time, we need year values for each row and also when you see in most of the values in date column only has ‘year’ value. So, let’s create a new column holding year values. Convert the column to 'datetime' type and extract the year to populate the new column. Display Top 5 records after the conversion.

```python
# Convert date column to datetime type
dataset['date'] = pd.to_datetime(dataset['date'])

# extract the year to populate the new 'year' column
dataset['year'] = pd.DatetimeIndex(dataset['date']).year

# Display Top 5 records after the conversion
dataset.head()
```

### Handling Missing Values

The column such as SO2, NO2, rspm, spm, pm2_5 are the ones which contribute much to our analysis. So, we need to remove null from those columns to avoid inaccuracy in the prediction.
We use the Imputer from sklearn.preprocessing to fill the missing values in every column with the mean.

```python
# define columns of importance, which shall be used reguarly (COLS = ....)
COLS = dataset[['so2', 'no2', 'rspm', 'spm', 'pm2_5']]
# invoke SimpleImputer to fill missing values using 'mean' as the replacement strategy
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imp.fit(COLS)

dataset[['so2', 'no2', 'rspm', 'spm', 'pm2_5']] = imp.transform(COLS)

dataset.head()
```

```python
# Display data.info after the transformation
dataset.info()
# Display that there are no more missing values in the dataset
dataset.isnull().sum()
```

### What is the yearly trend in a particular state, say ‘Andhra Pradesh’?

Create a new dataframe containing the NO2, SO2, rspm, and spm data regarding state ‘Andhra Pradesh’ only and group it by ‘year’. Display top 5 records after.

```python
# Create new dataframe

state_df = dataset[dataset.state == 'Andhra Pradesh']

andhra_pradesh_data = state_df[['year','no2','so2','rspm','spm']].groupby(by='year')

# Display top 5 records after
andhra_pradesh_data.head()
```

```python
# Display yearly trend graph (year vs. value) in pairs: (a) so2 and no2 (b) rspm and spm. 
# So, you will display TWO graphs altogether.

# so2 and no2 yearly trend graph
so2_pivot_table = state_df.pivot_table(values='so2', index='year', aggfunc='mean')
no2_pivot_table = state_df.pivot_table(values='no2', index='year', aggfunc='mean')

plt.plot(so2_pivot_table.index, so2_pivot_table.so2, linestyle = '-.', color='b', label='so2')
plt.plot(no2_pivot_table.index, no2_pivot_table.no2, linestyle = '--', color='r', label='no2')

plt.title('so2 and no2 Yearly Trends')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Value')
plt.show()
```

```python
# rspm and spm yearly trend graph

rspm_pivot_table = state_df.pivot_table(values='rspm', index='year', aggfunc='mean')
spm_pivot_table = state_df.pivot_table(values='spm', index='year', aggfunc='mean')

plt.plot(rspm_pivot_table.index, so2_pivot_table.so2, linestyle = '-.', color='b', label='rspm')
plt.plot(no2_pivot_table.index, no2_pivot_table.no2, linestyle = '--', color='r', label='spm')

plt.title('rspm and spm Yearly Trends')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Value')
plt.show()
```
